```python
import os
from PIL import Image
import matplotlib.pyplot as plt

root = "../data/images"

classes = os.listdir(root)
print("Classes:", classes[:5])

sample_class = classes[0]
img_path = os.path.join(root, sample_class, os.listdir(os.path.join(root, sample_class))[0])

img = Image.open(img_path)
plt.imshow(img)
plt.axis("off")
plt.show()
```
